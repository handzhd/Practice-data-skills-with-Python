## Calculate the average rating of apps

### First method to calculate the average of data by convert it to list

```python
from csv import reader

opened_file = open('AppleStore.csv', encoding="utf8")
read_file = reader(opened_file)
print(read_file)
apps_data = list(read_file)
print(apps_data[:2])
```

```python
rating_sum = 0
for row in apps_data[1:]:
    rating = float(row[8])
    rating_sum += rating
avg_rating = rating_sum / 7197
print(avg_rating)
```

## Separate the free and paid apps ratings

### Using first method from previous
#### but first I want to remove unnecessary columns in my dataset

```python
data_apps = apps_data
[i.pop(0) for i in data_apps]
print(data_apps[:2])
```

```python
free_rating_sum = []
paid_rating_sum = []
for row in data_apps[1:]:
    price = float(row[4])
    rating = float(row[7])
    if price == 0.0:
        free_rating_sum.append(rating)
    elif price != 0.0:
        paid_rating_sum.append(rating)


avg_rating_free = sum(free_rating_sum) / len(free_rating_sum)
print("The average of free apps: ", avg_rating_free)

avg_rating_paid = sum(paid_rating_sum) / len(paid_rating_sum)
print("The average of paid apps: ", avg_rating_paid)
    
```

### More exploring data
#### I want to see the average rating of games, social media, shopping, finance, music, and productivity apps
#### I want to create a new dataframe to store the rating of each genres apps

```python
apps_rating = []
genres = ['Games', 'Social Networking', 'Shopping', 'Finance', 'Music', 'Productivity']
Games = []
Social_Media = []
Shopping = []
Finance = []
Music = []
Productivity = []

for row in data_apps[1:]:
    rating = float(row[7])
    genre = row[11]
    if genre == genres[0]:
        Games.append(rating)
    elif genre == genres[1]:
        Social_Media.append(rating)
    elif genre == genres[2]:
        Shopping.append(rating)
    elif genre == genres[3]:
        Finance.append(rating)
    elif genre == genres[4]:
        Music.append(rating)
    elif genre == genres[5]:
        Productivity.append(rating)

games_avg_rating = sum(Games)/len(Games)
social_media_avg_rating = sum(Social_Media)/len(Social_Media)
shopping_avg_rating = sum(Shopping)/len(Shopping)
finance_avg_rating = sum(Finance)/len(Finance)
music_avg_rating = sum(Music)/len(Music)
productivity_avg_rating = sum(Productivity)/len(Productivity)

apps_rating.append(games_avg_rating)
apps_rating.append(social_media_avg_rating)
apps_rating.append(shopping_avg_rating)
apps_rating.append(finance_avg_rating)
apps_rating.append(music_avg_rating)
apps_rating.append(productivity_avg_rating)


# print(apps_rating)
# rating_apps = [[x] for x in apps_rating]
# rating_apps = [genres] + rating_apps
genres_rating = pd.DataFrame([apps_rating], columns= genres)
print(genres_rating)
genres_rating
```
